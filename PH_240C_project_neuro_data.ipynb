```python
from random import sample
from nonlinear_learning import *
import pandas as pd
```

```python
import rpy2.robjects as robjects
robjects.r['load']("s5_ECoG_label.RData")
series_low = np.array(robjects.r['ofc_s05'][0][0])
series_high = np.array(robjects.r['ofc_s05'][0][1])
xs_low = np.mean(series_low[0:10,:,:],0)
xs_high = np.mean(series_high[0:10,:,:],0)
xs_low  = (xs_low - np.mean(xs_low,0))/np.std(xs_low,0)
xs_high  = (xs_high - np.mean(xs_high,0))/np.std(xs_high,0)
```

```python
torch.manual_seed(2) 
max_iter=20
lambda0=0.05
w_threshold=0.1
d = xs_low.shape[1]
model=NotearsMLP(dims=[d, 10, 1], bias=True)
W_low_t = notears_nonlinear(model, xs_low, lambda1=lambda0, lambda2=lambda0, w_threshold=w_threshold,max_iter=max_iter)
W_low = W_low_t.T
```

```python
node_label = []
node_edge_num_in = []
node_edge_num_out = []
for i in range(d):
    node_label.append(i+1)
    node_edge_num_in.append(np.sum(W_low[i,:]!=0))
    node_edge_num_out.append(np.sum(W_low[:,i]!=0))

data = {'node_label': node_label , 'node_edge_num_in': node_edge_num_in , 'node_edge_num_out': node_edge_num_out}
df = pd.DataFrame(data=data)
pd.set_option('display.max_rows', df.shape[0]+1)
df
```
